```python
import numpy as np
import scipy.stats as sts
import scipy.special as sc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=[1.41,9.97,0.28,13.87,11.61,10.93,16.21,15.67,3.62,27.56,5.27,23.70,12.16,20.59,20.01,14.54,0.73 ,8.01 ,-6.80,28.74,8.85 ,20.76,24.17,28.81,14.92,14.71,13.97,18.14,2.21 ,23.46,17.60,4.75 ,6.16 ,18.26,24.47,9.25 ,15.41,8.44 ,29.20,19.62,2.33 ,24.57,11.56,-3.44,12.47,26.37,7.46 ,11.27,14.94,4.26]
```

---
Вариационный ряд выборки:

```python
X=sorted(X)
print(X)
```

---
Размах выборки:

```python
w=max(X)-min(X)
w
```

---
Повторяющихся элементов выборки нет:

```python
len(set(X))
```

---
Гистограмма и полигон частот выборки:

```python
hist_params=plt.hist(X,bins=7,density=True)
ivl_mid=list(map(lambda ivl:(ivl[0]+ivl[1])/2,zip(hist_params[1][:-1],hist_params[1][1:])))
poly=plt.plot(ivl_mid,hist_params[0])
```

---
Частоты появления элементов выборки в каждом интервале:

```python
freqs=hist_params[0]*50*w/7
freqs
```

Относительные частоты:

```python
rel_freqs=hist_params[0]*w/7
rel_freqs
```

Центральные точки интервалов:

```python
ivl_mid
```

---
Эмпирическая фунцкия распределения негруппированной выборки:

```python
plt.grid()
plt.plot([-8,X[0]],[0,0],color='blue')
plt.plot([X[-1],30],[1,1],color='blue')
for i in range(0,49):
    plt.plot([X[i],X[i+1]],[(i+1)/50,(i+1)/50],color='blue')
```

---
Эмпирическая функция распределения группированной выборки:

```python
Fx_gr=plt.hist(X,bins=7,density=True,cumulative=True,align='right')
plt.clf()
plt.grid()
plt.plot([-6,ivl_mid[0]],[0,0],color='blue')
for z,h in zip(zip(Fx_gr[1][:-1]+18/7,Fx_gr[1][1:]+18/7),zip(Fx_gr[0],Fx_gr[0])):
    plt.plot([z[0],z[1]],[h[0],h[1]],color='blue')
```

---
Выборочные числовые характеристики негруппированной выборки:

- математическое ожидание

```python
Mx=np.mean(X)
Mx
```

- смещённая дисперсия

```python
Dx=np.var(X)
Dx
```

- несмещённая дисперсия

```python
np.var(X,ddof=1)
```

- медиана

```python
hx=np.median(X)
hx
```

---
Выборочные числовые характеристики группированной выборки:

- математическое ожидание

```python
Mx_gr=sum(z*n for z,n in zip(ivl_mid,freqs))/50
Mx_gr
```

- смещённая дисперсия

```python
Dx_gr=(sum(z*z*n for z,n in zip(ivl_mid,freqs))-50*Mx_gr*Mx_gr)/50
Dx_gr
```

- несмещённая дисперсия

```python
Dx_gr*50/49
```

- мода

```python
d=np.argmax(freqs)
hist_params[1][d]+(freqs[d]-freqs[d-1])/(2*freqs[d]-freqs[d-1]-freqs[d+1])*w/7
```

- медиана

```python
for i,ivl in zip(range(0,7),zip(hist_params[1][:-1],hist_params[1][1:])):
    if hx>ivl[0] and hx<ivl[1]:
        h=i
hist_params[1][h]+(50/2-sum(freqs[:h]))/freqs[h]*w/7
```
